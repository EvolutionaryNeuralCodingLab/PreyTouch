```python
types = {
    3: PySpin.CBooleanPtr,
    5: PySpin.CFloatPtr,
    str: PySpin.CStringPtr,
    9: PySpin.CEnumerationPtr,
    6: PySpin.CValuePtr
}
NaN = 'x'

class NodeException(Exception):
    pass


class NodeMapReader:
    def __init__(self, nodemap, output_dict):
        self.nodemap = nodemap
        self.output_dict = output_dict
    
    def store(self, name):
        self.output_dict.setdefault(name, []).append(self.getn(name).ToString())
        
    def getn(self, name):
        try:
            node = self.nodemap.GetNode(name)
            ntype = node.GetPrincipalInterfaceType()
            ptr = types.get(ntype)
            if not ptr:
                raise NodeException(f'Unknown interface type: {ntype}')
            return ptr(node)
        except NodeException as e:
            print(e)
            return NaN
        except Exception:
            return NaN
    
    def setn(self, name, value):
        ptr = self.getn(name)
        if ptr == NaN:
            print(f'Unable to set value {value} for {name}')
            return False
        
        entry = ptr.GetEntryByName(value)
        if not PySpin.IsAvailable(entry) or not PySpin.IsReadable(entry):
            print(f'entry {name} is not available for {name}')
            return False
        ptr.SetIntValue(entry.GetValue())
        return True
    
```

```python
d = {}

for cam in cam_list:
    nodemap_tldevice = cam.GetTLDeviceNodeMap()
    device_name = PySpin.CStringPtr(nodemap_tldevice.GetNode('DeviceModelName')).GetValue()
    if 'firefly' not in device_name.lower():
        continue
        
    d.setdefault('device_name', []).append(device_name)
    d.setdefault('device_id', []).append(PySpin.CStringPtr(nodemap_tldevice.GetNode('DeviceID')).GetValue())
    
    cam.Init()
    nodemap = cam.GetNodeMap()
    reader = NodeMapReader(nodemap, d)
    cam.TriggerSelector.SetValue(PySpin.TriggerSelector_FrameStart)
    cam.TriggerMode.SetValue(PySpin.TriggerMode_On)
    cam.AcquisitionFrameRateEnable.SetValue(True)
    cam.TriggerActivation.SetValue(PySpin.TriggerActivation_RisingEdge)
    
#     cam.UserSetSave()
    
    
    reader.store('DeviceVersion')
    reader.store('AcquisitionFrameRate')
    reader.store('AcquisitionFrameRateEnable')
    reader.store('TriggerSource')
    reader.store('TriggerMode')
    reader.store('TriggerSelector')
    reader.store('TriggerActivation')
    reader.store('LineSelector')
#     reader.store('AcquisitionResultingFrameRate')
    
    cam.DeInit()

df = pd.DataFrame(d)
print(df)
```
